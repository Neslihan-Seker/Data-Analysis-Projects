***2. Evden İşe Mesafe ve İşten Ayrılma Oranı***
***Hipotez:
Uzakta oturan çalışanların işten ayrılma oranı daha mı yüksek?***

```python
plt.figure(figsize=(8,6))
sns.boxplot(x = data["Attrition"], y = data["DistanceFromHome"])
plt.xlabel("İşten Ayrılma Durumu")
plt.ylabel("Ev - İş Mesafesi (km)")
plt.title("Evden İşe Mesafe ve İşten Ayrılma İlişkisi")

plt.show()
```

```python
data["Attrition_Numeric"] = data["Attrition"].apply(lambda x: 1 if x == "Yes" else 0)
# Korelasyon hesapla
corr_value = data["Attrition_Numeric"].corr(data["DistanceFromHome"])
print(f"Ev-İş Mesafesi ile İşten Ayrılma Korelasyonu: {corr_value:.2f}")
```

***3. Çalışma Ortamı Memnuniyeti, İş Katılımı ve İşten Ayrılma-
Hipotez:
Çalışma ortamı memnuniyeti ve iş katılımı düşük olanlar daha mı fazla işten ayrılıyor?***

```python
# İşten ayrılan ve ayrılmayanların ortalama iş tatmini
mean_job_satisfaction = data.groupby("Attrition")["JobSatisfaction"].mean()
print(mean_job_satisfaction)
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.barplot(x=data["JobSatisfaction"], y=data["Attrition"] == "Yes", estimator=lambda x: sum(x) / len(x))
plt.xlabel("İş Tatmini (1 = Düşük, 4 = Yüksek)")
plt.ylabel("İşten Ayrılma Oranı")
plt.title("İş Tatmini ve İşten Ayrılma İlişkisi")
plt.show()
from scipy.stats import chi2_contingency

# İş tatmini seviyelerine göre işten ayrılma tablosu
contingency_table = pd.crosstab(data["JobSatisfaction"], data["Attrition"])
chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"Khi-kare Testi P-Değeri: {p}")
if p < 0.05:
    print("İş tatmini ve işten ayrılma arasında anlamlı bir ilişki var!")
else:
    print("İş tatmini ve işten ayrılma arasında anlamlı bir ilişki yok.")
```
